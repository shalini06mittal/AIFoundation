```python
# Simple one neuron - One data point

# Input value  - area
x = 2
# Actual Value - price /Correct Output
y = 10 # 5

#intial weight
w = 1.0

# learning rate
lr = 0.1
prev_loss = float('inf')
for i in range(10):
    y_pred = w * x

    # Loss ( MSE - mean Squared Error)
    loss = (y - y_pred) ** 2

    # Gradient descent (derivative of loss w.r.t w) error * input * constant
    grad = -2 * x * (y - y_pred)

    # update weight
    w_new = w - lr * grad
    print(f'{i} => Weight : {w:.4f}, Prediction: {y_pred:.4f}, Loss: {loss:.4f} , Grad: {grad:.4f}, New Weight: {w_new:.4f}')
    w = w_new
    if abs(prev_loss - loss) < 0.0001 : # abs(grad)
      print(f'{i} : Stopped Loss ~ 0 ')
      break
    prev_loss = loss
```

```python
'''
Input layer - 2 features
Hidden layer 1 - 3 neurons ( luxury, compactness, density)
Hidden layer 2 - 2 neurons (interactions bet ween the features)
Output layer - 1 neuron ( regression or classification)

'''
import numpy as np
# Input
x = np.array([[1000,2]])

# layer 1
# Weights for 2 features and 3 neurons
W1 = np.array([
    [0.9, 0.3, -0.2],
    [0.5, 0.7, 0.1]
])

b1 = np.array([[10, 5, -3]])

# layer 2
# Weights for 3 neurons passing to  2 neurons
W2 = np.array([
    [0.8, 0.4],
    [0.5, 0.1],
    [-0.5, 0.3]
])

b2 = np.array([[2, -1]])

# layer 3 - Output Layer
# Weights for 2 neurons passing to  1 neurons
W3 = np.array([
    [1.5],
    [-1.5]
])

b3 = np.array([[0.5]])

# Activation functions (Relu)
def relu(x):
  return np.maximum(0, x)

# Forward Pass
# Layer 1

Z1 = np.dot(x, W1) + b1
A1 = relu(Z1)

# Layer 2
Z2 = np.dot(A1, W2) + b2
A2 = relu(Z2)

# Output
Z3 = np.dot(A2, W3) + b3
# No activation on the output layer

print(f'Input: {x}, Layer1: {Z1}, Layer2: {Z2}, Output: {Z3}')
```

```python
import numpy as np

# -----------------------------
# Sample dataset (area, bedrooms, distance)
# -----------------------------
X = np.array([
    [1000, 2, 5],
    [2000, 4, 2],
    [1500, 3, 3],
    [800,  2, 10],
    [2500, 5, 1]
])

# Labels: 0 = Budget, 1 = Premium
y = np.array([[0], [1], [1], [0], [1]])

# scaling , Normalize the data
X = X / np.max(X, axis=0)

# Initial Weights
# Training model => train and test split
np.random.seed(42) # Fixes randomness -> same results every run

W1 = np.random.randn(3,4)*0.1 # 3 inputs, 4 neurons
b1 = np.zeros((1,4))

W2 = np.random.randn(4,3)*0.1 # 4 inputs, 3 neurons, Guassian formula
b2 = np.zeros((1,3))

W3 = np.random.randn(3,2)*0.1 # 3 inputs, 2 neurons
b3 = np.zeros((1,2))

W4 = np.random.randn(2,1)*0.1 # 2 neurons -> 1 output
b4 = np.zeros((1,1))

# learning rate
lr = 0.5
```

```python
# Activation Functions
# -ve values
def relu(x):
  return np.maximum(0,x)

def derivative_relu(x):
  return (x>0).astype(float) # false 0.0 and true 1.0

def leaky_relu(x):
  return np.where(x>0, 1, 0.01 * x)

def leaky_relu_derivative(x):
    return np.where(x > 0, 1, 0.01)

def sigmoid(x):
  #return 1/(1+np.exp(-x))
  return 1 / (1 + np.exp(-np.clip(x, -500, 500)))
```

```python
# Training Loop

for epoch in range(5000):
  # layer 1 forward pass
  Z1 = X.dot(W1) + b1
  A1 = leaky_relu(Z1)

  Z2 = A1.dot(W2) + b2
  A2 = leaky_relu(Z2)

  Z3 = A2.dot(W3) + b3
  A3 = leaky_relu(Z3)

  Z4 = A3.dot(W4) + b4

  # predict
  y_pred = sigmoid(Z4)

  # Loss
  loss = np.mean((y - y_pred) ** 2)

  # print(loss, ':', y_pred)

  # STEP 1 : Calculate loss gradient
  # How much does the pre-activation of final layer affect the loss

  #dZ4 = (y_pred - y) * y_pred * ( 1- y_pred) # error x sensitivity

  # ---- Backprop (CORRECTED) ----
  # Output layer gradient
  dLoss_dyPred = 2 * (y_pred - y) / len(y)  # MSE derivative with mean
  dSigmoid = y_pred * (1 - y_pred)          # Sigmoid derivative
  dZ4 = dLoss_dyPred * dSigmoid             # Chain rule

  # (y_pred - y) derivateive of MSE
  # y_pred * ( 1- y_pred) derivative of sigmoid

  # Step 2: Calculate weight and bias gradients Layer 4
  # How much should we adjust W4 and B4 to reduce loss

  dW4 = A3.T.dot(dZ4) # gradient for weights = activation from previous layer x error signal. How much Layer 3 contribute to the error

  dB4 = np.sum(dZ4 , axis=0) # sum error across all samples so that bias affects equally

  # Step 3: Pass error backward to Layer 3
  # How much did each neuron in A3 contribute to the error

  dA3 = dZ4.dot(W4.T) # large weights = larger blame How muc error came from next layer

  # Step 4: Apply activation for Layer 3
  # Which neurons in Layer 3 actually contributed

  dZ3 = dA3 * leaky_relu_derivative(Z3) #this is gate if neuron active pass gradient, if 0 -> block


  # Repeat above steps for Layer 2, 1
  dW3 = A2.T.dot(dZ3)
  dB3 = np.sum(dZ3 , axis=0, keepdims=True)

  dA2 = dZ3.dot(W3.T)
  dZ2 = dA2 * leaky_relu_derivative(Z2)
  dW2 = A1.T.dot(dZ2)
  dB2 = np.sum(dZ2, axis = 0, keepdims=True)

  dA1 = dZ2.dot(W2.T)
  dZ1 = dA1 * leaky_relu_derivative(Z1)
  dW1 = X.T.dot(dZ1)
  dB1 = np.sum(dZ1, axis = 0, keepdims=True)

  # Update Weights
  W4 -= lr * dW4
  b4 -= lr * dB4

  W3 -= lr * dW3
  b3 -= lr * dB3

  W2 -= lr * dW2
  b2 -= lr * dB2


  W1 -= lr * dW1
  b1 -= lr * dB1

  if(epoch % 100 ==0):
    print(f'Epoch! {epoch} , Loss {loss}, Predicted {y_pred}')
```

```python
import tensorflow as tf

# Set seed
tf.random.set_seed(42)

# Bbuild model

model = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='leaky_relu', kernel_initializer='he_normal', input_shape=(3,)),
    tf.keras.layers.Dense(1, activation='sigmoid'),
])

# Define optimizer and loss
optimizer = tf.keras.optimizers.SGD(learning_rate=1.0)
loss = tf.keras.losses.BinaryCrossentropy()

model.compile(optimizer , loss, metrics =['accuracy'])

# train model

history = model.fit(X, y, epochs = 10000, verbose = 0)
```
